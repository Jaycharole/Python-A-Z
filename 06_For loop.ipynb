# Python for Loop

The for loop in Python is used to iterate over a sequence (list, tuple, string) or other iterable objects.

Iterating over a sequence is called traversal.

# Syntax

     for element in sequence:
         
         Body of for loop

Here, element is the variable that takes the value of the item inside the sequence on each iteration.

Loop continues until we reach the last item in the sequence.

# Example

# Find product of all element present in the list

```python
list = [10, 20, 30, 40, 50]

product = 1
for i in list:
    product = product * i
    
print(f"Product of list is {product}")
```

# range() function

We can generate a sequence of numbers using range() function. range(10) will generate numbers from 0 to 9 (10 numbers).

We can also define the start, stop and step size as range(start,stop,step size). step size defaults to 1 if not provided.

This function does not store all the values in memory, it would be inefficient. So it remembers the start, stop, step size and generates the next number on the go.

# Syntax of range() function

    range(start,stop,step size) 
    
- The start represents the beginning of the iteration.
- The stop represents that the loop will iterate till stop-1. The range(1,5) will generate numbers 1 to 4 iterations. It is optional.
- The step size is used to skip the specific numbers from the iteration. It is optional to use. By default, the step size is 1. It is optional.

```python
#print range of 10

for i in range(10):
    print(i)
```

```python
#print range of numbers from 1 to 20 with step size 2
for i in range(1,20,2):
    print(i)
```

```python
list = ["satish", "srinu", "murali", "naveen", "bramha"]

#iterate over the list using index
for i in range(len(list)):
    print(list[i])
```

```python
for i in list:
    print(i)
```

# for loop with else

A for loop can have an optional else block as well. The else part is executed if the items in the sequence used in for loop exhausts.

break statement can be used to stop a for loop. In such case, the else part is ignored.

Hence, the for loop else runs if no break occurs.

```python
list = [1, 2, 3]

for i in list:
    print(i)

else:
    print("No elements left in the list")
```

```python
list = [1, 3, 5, 7, 9, 10, 2]

for i in list:
    print(i)
    if i % 2 == 0:
        break
else:
    print("No items left in the list")
```

# Python program to display all prime numbers within a range

```python
range1 = int(input("Enter the start range: "))
range2 = int(input("Enter the end range: "))

print(f"Prime numbers between {range1} and {range2} are")

for num in range(range1,range2+1):
    if num > 1:
        isDivisible = False;
        for index in range(2, num):
            if num % index == 0:
                isDivisible = True;
        if not isDivisible:
            print(num);


```

```python
#Program to print table of given number.
num = int(input("Enter the number:"))
print(f'{num} table is as follows:')

for i in range(1,11):
    t = num * i
    print(f'{num} * {i} = {t}')
```
